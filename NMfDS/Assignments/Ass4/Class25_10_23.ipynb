```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Create n (x,y) coordinates spaced around a circle of radius r
def circle_init_conds(r, n):
    # Create evenly spaced angles
    theta = np.linspace(0, 2*np.pi, n)
    # Create x and y coordinates
    x = r*np.cos(theta)
    y = r*np.sin(theta)
    # Return the coordinates
    return np.array(x), np.array(y)

r = 10**-6
n = 100

# Plot circle
x, y = circle_init_conds(r, n)
plt.plot(x, y, 'o')
```

```python
from scipy.integrate import solve_ivp

r = 10**-6
n = 100

# Vector field
# x_1' = 8x_1
# x_2' = 2x_2

def vector_field(t, X):
    x_1, x_2 = X
    return [8*x_1, 2*x_2]

# Integration points
t_max = 10
t = [0, t_max]

x, y = circle_init_conds(r, n)
# Plot the linear system with the initail conditions, by integrating the system over time
for i in range(n):
    x_0 = x[i]
    y_0 = y[i]
    # Now integrate the system using solve_ivp
    sol = solve_ivp(vector_field, t, [x_0, y_0],t_eval=np.linspace(t[0], 
                                                             t[1], 100))
    # Plot the solution
    plt.plot(sol.y[0], sol.y[1], 'b')

plt.plot(x, y, 'o')
# Fix axis to be square, -20 to 20
plt.axis([-20, 20, -20, 20])
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
# Making it square
plt.gca().set_aspect('equal', adjustable='box')
plt.show()
plt.axis([-20, 20, -20, 20])
plt.plot(x, y, 'o')
# Fix axis to be square, -20 to 20
plt.axis([-20**20, 20**20, -20, 20])
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
# Making it square
plt.gca().set_aspect('equal', adjustable='box')
plt.show()
```
